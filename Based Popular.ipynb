```python
def imdb_score(df, q=0.9):
    df = df.copy()
    
    m = df['votes'].quantile(q) # syarat minimum vote
    C = (df['rating'] * df['votes']).sum() / df['votes'].sum() # rata-rata dari semua film
    
    df = df[df['votes'] >= m] # filter berdasarkan minimum vote
    df['score'] = df.apply(lambda x: (x.rating * x.votes + C*m) / (x.votes + m), axis=1)
    return df
```

```python
df = imdb_score(df)
```

### ML Engineering

```python
class RecomenderSystem:
    def __init__(self, data):
        self.df = pd.read_csv(data)
    
    def recommend(self, types_menu=None, topk=10):
        df = self.df.copy()
        df = self.demographic_filter(df, types_menu=types_menu)
        df = self.compute_imdb_score(df)
        
        result = df.loc[:, "restaurant":"rating"] 
        result = result.sort_values("rating", ascending=False)
        result = result.head(topk) 
        return result
    
    @staticmethod 
    def demographic_filter(df, types_menu=None, topk=10):
        df = df.copy()
        
        if types_menu is not None :
            df = df[(df['types of menu'] == types_menu)]
        return df
        
        
    @staticmethod
    def compute_imdb_score(df, q=0.9):
        df = df.copy()
    
        m = df['votes'].quantile(q) # syarat minimum vote
        C = (df['rating'] * df['votes']).sum() / df['votes'].sum() # rata-rata dari semua film

        df = df[df['votes'] >= m] # filter berdasarkan minimum vote
        df['score'] = df.apply(lambda x: (x.rating * x.votes + C*m) / (x.votes + m), axis=1)
        return df
```

```python
recsys = RecomenderSystem(data='data/df_zomato.csv')
```

```python
recsys.recommend()
```
